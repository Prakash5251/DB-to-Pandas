```python
## importing required python libraries

import pandas as pd
import mysql.connector
from sqlalchemy import create_engine
```

##### 
> SQLAlchemy is the Python SQL toolkit that allows developers to access and manage SQL databases using Pythonic domain language. 
> It provides a “Pythonic” way of interacting with those databases.

### we dont have to create connection separately if we are using sqlalchemy.

#### 

creating connection with mysql db and providing my credentials to access the db tables

conn = mysql.connector.connect(host = "localhost",
                               user = "root",
                               password = "12345678",
                               database = "test")

```python
# create engine to connect with MYSQL

engine = create_engine('mysql+mysqlconnector://root:12345678@localhost/test')
```

```python
## writing a query to fetch data from db table
query = 'select * from test.Account'
```

```python
## create a dataframe to render data in dataframe, and this time we will use engine instead of conn.
df = pd.read_sql(query, engine)
df
```

```python
## success now i can see the data
```

### Now lets see how we can insert data from pandas to mysql.

```python
data = [(15,"ACC170929","Sonu Singh", "19901","Savings")]
df_new = pd.DataFrame(data, columns=["account_id","account_number","account_name","balance","account_type"])
df_new
```

```python
## pushing the dataframe to mysql

df_new.to_sql("Account", engine, if_exists = "append", index=False)
```
