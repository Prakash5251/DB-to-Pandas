### Read entire table in a dataframe using read_sql_table

```python
df = pd.read_sql_table('Account',engine)
df
```

### If want to read only selected columns from table in a dataframe using read_sql_table

```python
df = pd.read_sql_table('Account',engine,columns= ["account_number","account_name", "balance"])
df
```

#### Join two tables and read them in a dataframe using read_sql_query

#### query = '''
 SELECT customers.name, customers.phone_number, orders.name, orders.amount
 FROM customers INNER JOIN orders
 ON customers.id=orders.customer_id
'''
df = pd.read_sql_query(query,engine)
df

### Query from tables and read them in a dataframe using read_sql_query

```python
query = """select * from Account"""

pd.read_sql_query(query,engine)
```

```python
query2 = """SELECT account_type, COUNT(*) AS count, SUM(balance) AS total_balance
FROM Account
GROUP BY account_type;"""

df2 = pd.read_sql_query(query2,engine)
df2
```

```python
df2
```

### read_sql is a wrapper around read_sql_query and read_sql_table 

```python
query3 = """SELECT account_type, COUNT(*) AS count, SUM(balance) AS total_balance
FROM Account
GROUP BY account_type;"""

df3 = pd.read_sql(query2,engine)
df3
```

```python
pd.read_sql("Account",engine)
```

### Write to mysql database using to_sql

```python
df5 = pd.read_csv("/Users/satya.ps/Downloads/mobile.csv")
df5
```

```python
df6 = pd.read_csv("mobile.csv")
df6
```

#### Rename the pivots of dataframe

```python
df6.rename(columns={
    'cust_name': 'customer_name',
    'phone': 'mobile'
}, inplace = True)
```

```python
df6
```

```python
## Write to mysql database using to_sql

df5.to_sql(
    name='mobile', # database table name
    con=engine,
    if_exists='append',
    index=False
)
```

```python
## Now read the table data to validate if we Inserted any in table

pd.read_sql("mobile",engine)
```

##### to_sql has different parameters such as chunksize which allows to write data in chunks. This is useful when size of dataframe is huge
